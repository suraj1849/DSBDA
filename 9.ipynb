```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/titanic_data.csv")
```

```python
df
```

```python
sns.boxplot(x="Sex",y="Age",hue="Survived",data=df,palette="Set2")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.title("Age distribution with gender and survival status")
plt.legend(title="Survived",loc="upper right")
plt.show()
```
